```python
import matplotlib.pyplot as plt
```

```python
def create_image(rgb_img,h1,w1,h2,w2,ans):
    crop_img111 = "crop_img" + "009_" + ans
    crop_img1 = crop_img111 + ".png"
    crop_img11 = rgb_img[h1:h2, w1:w2]
    plt.figure()
    plt.imshow(crop_img11)
    #print(crop_img11.shape)
    cv2.imwrite(crop_img1, crop_img11)
```

```python
import cv2
```

```python
img = cv2.imread("./diaretdb1_v_1_1/resources/images/ddb1_fundusimages/image009.png")
```

```python
#cv2.imshow('image',img)
b,g,r = cv2.split(img)       # get b,g,r
rgb_img = cv2.merge([r,g,b])
r_img = cv2.merge([r])
# switch it to rgb
```
